# LAB #2 Solution
# Using crosstab () over data frame 'orders_data' calculate:
# STEP 1: Average sales value for 'Province' and 'Product Category' columns
# STEP 2: Sales amount for 'Province' and 'Product Category' columns

```python
import pandas as pd
orders_data = pd.read_excel("orders.xls")
```

```python
# STEP 1
pd.crosstab(orders_data['Province'],orders_data['Product Category'], 
            margins=True, values=orders_data['Sales'], aggfunc='mean')
```

```python
# STEP 2
pd.crosstab(orders_data['Province'],orders_data['Product Category'], 
            margins=True, values=orders_data['Sales'], aggfunc='sum')
```
