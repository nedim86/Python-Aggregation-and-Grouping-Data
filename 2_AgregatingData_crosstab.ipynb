# Agregating Data - crosstab()

```python
import pandas as pd
data_BM=pd.read_csv('big_mart_sales.csv')
# Replace missing values
data_BM.loc[(data_BM.Item_Weight.isna() == True) , 'Item_Weight'] = round(data_BM.Item_Weight.mean(),3)
data_BM.loc[(data_BM.Outlet_Size.isna() == True) , 'Outlet_Size'] = 'Medium'
data_BM
```

```python
# Total number of rows and columns of data frame data_BM
data_BM.shape
```

# EXAMPLE 1 ( crosstab() )

```python
# Apply crosstab() over the 'Outlet_Size' and 'Outlet_Location_Type' columns
# The default aggregate function for crosstab() is count (counts rows for columns defined in crosstab)
# margins = True - adds summed values for columns and rows
pd.crosstab(data_BM['Outlet_Size'], data_BM['Outlet_Location_Type'],
            margins=True, margins_name="Total")
```

```python
# Apply crostab () without margins = True and margins_name
pd.crosstab(data_BM['Outlet_Size'],data_BM['Outlet_Location_Type'])
```

# EXAMPLE 2 ( crosstab() )

```python
# Overview of the initial data frame data_BM
data_BM
```

```python
# We want to calculate the number (frequency) of products per group 'Item_Type' in each group 'Outlet_Size'
pd.crosstab(data_BM['Outlet_Size'],data_BM['Item_Type'], margins=True, margins_name="Total")
```

```python
# We want to calculate the number (frequency) of products per group 'Outlet_Size' in each group 'Item_Type' 
# Identical results as in the previous example, only a different display of rows and columns
pd.crosstab(data_BM['Item_Type'], data_BM['Outlet_Size'], margins=True, margins_name="Total")
```

# EXAMPLE 3 ( crosstab() )

```python
# Load data from the new orders.xls file into the data frame orders_data
orders_data = pd.read_excel("orders.xls")
```

```python
orders_data
```

```python
# Total number of rows and columns of data frame orders_data
orders_data.shape
```

# EXAMPLE 3.1.

```python
# Apply crosstab () by columns 'Province' and 'Customer Segment'
pd.crosstab(orders_data['Province'], orders_data['Customer Segment'], 
            margins=True)
```

# EXAMPLE 3.2.

```python
# If we define the aggfunc attribute, we must also define the values attribute (The default aggregate function is countt
# We will count all OrderIDs, for each 'Customer Segment' in each 'Province'
pd.crosstab(orders_data['Province'], orders_data['Customer Segment'], 
            margins=True, values=orders_data['OrderID'], aggfunc='count')
```

# EXAMPLE 3.3.

```python
# We will caluculate the average 'Sales' for each 'Customer Segment' in each 'Province'
pd.crosstab(orders_data['Province'], orders_data['Customer Segment'], 
            margins=True, values=orders_data['Sales'], aggfunc='mean')
```
