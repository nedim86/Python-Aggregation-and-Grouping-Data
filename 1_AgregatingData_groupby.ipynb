# Agregating Data - groupby()

```python
import pandas as pd
data_BM=pd.read_csv('big_mart_sales.csv')
# Replace missing values
data_BM.loc[(data_BM.Item_Weight.isna() == True) , 'Item_Weight'] = round(data_BM.Item_Weight.mean(),3)
data_BM.loc[(data_BM.Outlet_Size.isna() == True) , 'Outlet_Size'] = 'Medium'
data_BM
```

```python
# Total number of rows and columns of data frame data_BM
data_BM.shape
```

# EXAMPLE 1 (Grouping by one column)

```python
# Group data by column 'Item_Type'
price_by_item = data_BM.groupby('Item_Type')
price_by_item.first()
```

```python
# If we want to display data from only one group, we use get_group
price_by_item.get_group('Breakfast') 
```

```python
# If we want to display data from only one group, we use get_group
price_by_item.get_group('Meat') 
```

```python
# We want to calculate the mean of the 'Item_MRP' column for each group
price_by_item.Item_MRP.mean()
```

```python
# We want to calculate the amount for the 'Item_Outlet_Sales' column for each group
price_by_item.Item_Outlet_Sales.sum()
```

```python
# We want to calculate the number of products for each group
price_by_item.Item_Identifier.count()
```

```python
# We want to calculate the maximum product weight for each individual group
price_by_item.Item_Weight.max()
```

```python
# We want to calculate the minimum product weight for each individual group
price_by_item.Item_MRP.min()
```

# EXAMPLE 2 (Average sales by year)

```python
# Overview of the initial data frame data_BM
data_BM
```

# EXAMPLE 4 (Preparatory actions before grouping)

```python
# Group by multiple columns 'Item_Type', 'Item_Fat_Content'
multiple_groups = data_BM.groupby(['Item_Type','Item_Fat_Content'])

# We will show a few rows of grouped data frame
# Problem in subgroups 'Item_Fat_Content' (we need to modify the data first)
multiple_groups.first()
```

```python
# Previous grouping is not OK (Low Fat, LF, low fat is one group, not three)
# Let's do preparatory work
mapping = {
    'Low Fat' : 'LF',
    'Regular' : 'R',
    'LF' : 'LF',
    'reg': 'R',
    'low fat' : 'LF'
}
data_BM.Item_Fat_Content = data_BM.Item_Fat_Content.map(mapping)
```

```python
# Group by multiple columns 'Item_Type', 'Item_Fat_Content'
multiple_groups = data_BM.groupby(['Item_Type','Item_Fat_Content'])

# let's show a few first lines
multiple_groups.first()
```

```python
multiple_groups.Item_MRP.mean()
```

```python
multiple_groups.Item_Outlet_Sales.mean()
```
